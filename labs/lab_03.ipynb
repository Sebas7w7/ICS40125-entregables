```python
import pandas as pd

# Cargar datos (movido aquí para asegurar que df esté definido)
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/netflix_titles.csv')

print(f"El dataset tiene {df.shape[0]} filas y {df.shape[1]} columnas.")
print("\nTipos de datos y valores no nulos:")
df.info()
print("\nValores nulos por columna:")
print(df.isnull().sum())

# Transformar 'date_added' a tipo fecha, manejando formatos mixtos
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
print("\n'date_added' después de la transformación:")
print(df['date_added'].head())

# Crear columnas auxiliares con assign
df = df.assign(
    year_added=df['date_added'].dt.year,
    month_added=df['date_added'].dt.month
)

print("\nDataFrame con nuevas columnas 'year_added' y 'month_added':")
display(df.head())
```

```python
# 4. Seleccionar películas añadidas después de 2018
movies_after_2018 = df.loc[(df['type'] == 'Movie') & (df['year_added'] > 2018)]
print("\nPelículas añadidas después de 2018:")
display(movies_after_2018.head())
print(f"Número de películas añadidas después de 2018: {len(movies_after_2018)}")

# 5a. Filtrar títulos que contienen 'love' (insensible a mayúsculas/minúsculas)
love_titles = df.loc[df['title'].str.contains('love', case=False, na=False)]
print("\nTítulos que contienen 'love':")
display(love_titles[['title', 'description']].head())
print(f"Número de títulos que contienen 'love': {len(love_titles)}")

# 5b. Extraer la duración en minutos para películas
# Primero, asegurarnos de trabajar solo con 'Movie' para la duración en minutos
movies_df = df.loc[df['type'] == 'Movie'].copy()
# Fix: Use raw string for regex to avoid SyntaxWarning
movies_df['duration_minutes'] = movies_df['duration'].str.extract(r'(\d+)\smin', expand=False).astype(float)
print("\nPelículas con duración en minutos extraída:")
display(movies_df[['title', 'duration', 'duration_minutes']].head())
```

```python
# 6. Aplicar explode() sobre la columna 'listed_in' para obtener una fila por cada género.
df_exploded_genres = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')
print("\nDataFrame con 'listed_in' expandido (primeras filas):")
display(df_exploded_genres.head())

# 7. Obtener un top 10 de géneros más frecuentes utilizando value_counts().
top_10_genres = df_exploded_genres['listed_in'].value_counts().head(10)
print("\nTop 10 géneros más frecuentes:")
print(top_10_genres)

# 8. Aplicar where() y mask() para marcar las películas de más de 120 minutos como contenido largo.
# Aseguramos que 'duration_minutes' esté disponible, si no, lo creamos para el dataframe completo
# Esto es redundante si movies_df ya existe y fue calculado, pero asegura la columna en el df principal
if 'duration_minutes' not in df.columns:
    df['duration_minutes'] = df['duration'].str.extract(r'(\d+)\smin', expand=False).astype(float)

df['content_type'] = 'Corto'
df['content_type'] = df['content_type'].where(df['duration_minutes'] <= 120, 'Largo')
df['content_type'] = df['content_type'].mask(df['duration_minutes'].isnull(), 'Desconocido')
print("\nClasificación de contenido (Corto/Largo) para algunas películas:")
display(df[['title', 'duration_minutes', 'content_type']].sample(5))

# 9. Utilizar .loc para filtrar películas que cumplen con:
#    Más de 100 minutos de duración.
#    Rating igual a 'R'.
#    País igual a 'United States'.
filtered_movies = df.loc[
    (df['duration_minutes'] > 100) &
    (df['rating'] == 'R') &
    (df['country'] == 'United States')
]
print("\nPelículas filtradas (más de 100 min, Rating 'R', United States):")
display(filtered_movies[['title', 'duration_minutes', 'rating', 'country']].head())
print(f"Número de películas que cumplen los criterios: {len(filtered_movies)}")

# 10. Utilizar .style para formatear visualmente el top 10 de películas más largas.
# Aseguramos que movies_df ya contiene 'duration_minutes'
# Si movies_df no tiene 'duration_minutes' por alguna razón (e.g. celda anterior no ejecutada)
# la generamos aquí de nuevo para el contexto de este paso.
if 'duration_minutes' not in movies_df.columns:
    movies_df['duration_minutes'] = movies_df['duration'].str.extract(r'(\d+)\smin', expand=False).astype(float)

top_10_longest_movies = movies_df.sort_values(by='duration_minutes', ascending=False).head(10)

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

styled_top_10 = top_10_longest_movies[['title', 'duration_minutes', 'release_year', 'rating']].style \
    .format({'duration_minutes': "{:.0f} min"}) \
    .bar(subset=['duration_minutes'], color='lightblue') \
    .apply(highlight_max, subset=['duration_minutes'])

print("\nTop 10 películas más largas (formato visual):")
display(styled_top_10)
```
