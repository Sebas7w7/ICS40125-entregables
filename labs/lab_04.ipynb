A partir de los archivos disponibles, realice los siguientes pasos:

**a)** Cree un DataFrame llamado `df_anio` que consolide la información proveniente de los archivos **`libertad_prensa_01.csv`** y **`libertad_prensa_02.csv`**, correspondientes a distintas ventanas de tiempo. Recuerde que ambos archivos tienen nombres de columnas en mayúscula, por lo que debe normalizarlas a **minúscula** para asegurar consistencia.

**b)** Explore el archivo **`libertad_prensa_codigo.csv`** e identifique el código ISO que aparece asociado a dos nombres de país distintos. Elimine el registro que corresponda a un valor incorrecto o inconsistente, conservando solo el que considere válido.

**c)** Una vez preparados los archivos, cree un nuevo DataFrame llamado `df` que combine `df_anio` con `df_codigos`, utilizando la columna `codigo_iso` como clave. Asegúrese de realizar una unión que conserve únicamente los registros que tengan coincidencia en ambas fuentes.

> * Para unir los archivos por filas (años), utilice la función `pd.concat([...])`.
> * Para combinar información por columnas (variables), utilice `pd.merge(...)` especificando `on='codigo_iso'`.



````python
import numpy as np
import pandas as pd

# Lectura de datos (se incluyó aquí para evitar errores de variable no definida)
archivos_anio = [
    'https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/libertad_prensa_01.csv',
    'https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/libertad_prensa_02.csv'
 ]
df_codigos = pd.read_csv('https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/libertad_prensa_codigo.csv')

df_list = []
for file_url in archivos_anio:
    df_temp = pd.read_csv(file_url)
    df_temp.columns = df_temp.columns.str.lower() # Normalizar nombres de columnas a minúsculas
    df_list.append(df_temp)

df_anio = pd.concat(df_list, ignore_index=True)

# El código ISO 'FRA' está asociado a dos nombres: 'France' y 'Francia'.
# Mantendré 'Francia' para mantener la consistencia del idioma.
df_codigos = df_codigos[df_codigos['pais'] != 'France'].reset_index(drop=True)

# Combinar df_anio con df_codigos
df = pd.merge(df_anio, df_codigos, on='codigo_iso', how='inner')

display(df.head())
````

* El país con el menor valor de `indice` (mayor libertad de prensa).
* El país con el mayor valor de `indice` (menor libertad de prensa).

**b)** Resuelve la misma tarea del punto anterior utilizando un enfoque vectorizado con `groupby`, sin usar ciclos explícitos.



#### **Lista de países latinoamericanos considerada:**

```python
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
           'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
```

````python
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

df_america = df[df['codigo_iso'].isin(america)].copy()

print("\n--- Análisis por año (usando ciclo for) ---")
for anio in df_america['anio'].unique():
    # Filtrar valores nulos en 'indice' para evitar errores con idxmin/idxmax
    df_anio_filtered = df_america[df_america['anio'] == anio].dropna(subset=['indice'])

    if not df_anio_filtered.empty:
        # País con menor índice (mayor libertad de prensa)
        min_indice_row = df_anio_filtered.loc[df_anio_filtered['indice'].idxmin()]
        print(f"Año {anio}: Mayor libertad de prensa - {min_indice_row['pais']} (Índice: {min_indice_row['indice']:.2f})")

        # País con mayor índice (menor libertad de prensa)
        max_indice_row = df_anio_filtered.loc[df_anio_filtered['indice'].idxmax()]
        print(f"Año {anio}: Menor libertad de prensa - {max_indice_row['pais']} (Índice: {max_indice_row['indice']:.2f})")
    else:
        print(f"Año {anio}: No hay datos de 'indice' disponibles para este año.")

print("\n--- Análisis por año (enfoque vectorizado con groupby) ---")
# Eliminar valores NaN antes de usar idxmin/idxmax en el enfoque vectorizado
df_america_cleaned = df_america.dropna(subset=['indice'])

# País con menor índice por año
min_indice_por_anio = df_america_cleaned.loc[df_america_cleaned.groupby('anio')['indice'].idxmin()]
print("Países con mayor libertad de prensa por año (menor índice):\n", min_indice_por_anio[['anio', 'pais', 'indice']].reset_index(drop=True))

# País con mayor índice por año
max_indice_por_anio = df_america_cleaned.loc[df_america_cleaned.groupby('anio')['indice'].idxmax()]
print("\nPaíses con menor libertad de prensa por año (mayor índice):\n", max_indice_por_anio[['anio', 'pais', 'indice']].reset_index(drop=True))
````

### 4. Análisis anual del índice por país

En esta sección se busca analizar la evolución del **índice máximo** de libertad de prensa alcanzado por cada país a lo largo del tiempo.

#### **Tarea principal:**

* Construye una tabla dinámica (`pivot_table`) donde las **filas** correspondan a los países, las **columnas** a los años (`anio`) y los **valores** sean el `indice` máximo alcanzado por cada país en ese año.
* Asegúrate de reemplazar los valores nulos resultantes con `0`.

> **Hint**: Puedes utilizar el parámetro `fill_value=0` en `pd.pivot_table(...)`.



#### **Preguntas adicionales:**

**a)** ¿Qué país tiene el mayor valor de `indice` en toda la tabla resultante? ¿Y cuál tiene el menor (distinto de cero)?
**b)** ¿Qué años presentan en promedio los valores de `indice` más altos? ¿Y los más bajos?

> (Pista: usa `.mean(axis=0)` sobre la tabla pivot)

**c)** ¿Qué país muestra mayor **variabilidad** (diferencia entre su máximo y mínimo `indice` a lo largo del tiempo)?

> (Pista: aplica `.max(axis=1) - .min(axis=1)`)

**d)** ¿Existen países con índice constante a lo largo de todos los años registrados? ¿Cuáles?

**e)** ¿Qué países no tienen ningún dato (es decir, quedaron con todos los valores igual a 0)? ¿Podrías explicar por qué?





````python
# Construir la tabla dinámica
pivot_table_df = pd.pivot_table(df, values='indice', index='pais', columns='anio', aggfunc='max', fill_value=0)

print("--- Tabla Dinámica del Índice de Libertad de Prensa (Máximo por País y Año) ---")
display(pivot_table_df.head())

# Pregunta a) ¿Qué país tiene el mayor valor de indice en toda la tabla resultante? ¿Y cuál tiene el menor (distinto de cero)?
max_indice_global = pivot_table_df.max().max()
country_max_idx = pivot_table_df.stack().idxmax()
print(f"\nPaís con el mayor valor de 'indice' en toda la tabla: {country_max_idx[0]} en el año {country_max_idx[1]} con un índice de {max_indice_global}")

# Para el menor valor distinto de cero
min_indice_nonzero = pivot_table_df.replace(0, np.nan).min().min()
country_min_idx = pivot_table_df.replace(0, np.nan).stack().idxmin()
print(f"País con el menor valor de 'indice' (distinto de cero) en toda la tabla: {country_min_idx[0]} en el año {country_min_idx[1]} con un índice de {min_indice_nonzero}")

# Pregunta b) ¿Qué años presentan en promedio los valores de indice más altos? ¿Y los más bajos?
average_indice_by_year = pivot_table_df.mean(axis=0)
year_max_avg = average_indice_by_year.idxmax()
year_min_avg = average_indice_by_year.idxmin()
print(f"\nAño con el promedio de 'indice' más alto: {year_max_avg} (Promedio: {average_indice_by_year.max():.2f})")
print(f"Año con el promedio de 'indice' más bajo: {year_min_avg} (Promedio: {average_indice_by_year.min():.2f})")

# Pregunta c) ¿Qué país muestra mayor variabilidad (diferencia entre su máximo y mínimo indice a lo largo del tiempo)?
variability = pivot_table_df.replace(0, np.nan).max(axis=1) - pivot_table_df.replace(0, np.nan).min(axis=1)
country_most_variable = variability.idxmax()
print(f"\nPaís con mayor variabilidad en el 'indice': {country_most_variable} (Diferencia: {variability.max():.2f})")

# Pregunta d) ¿Existen países con índice constante a lo largo de todos los años registrados? ¿Cuáles?
# Filter out countries that have 0 for some years (meaning no data or 0 index)
non_zero_pivot_table = pivot_table_df.replace(0, np.nan)

# Calculate max and min for each country, ignoring NaNs
country_max = non_zero_pivot_table.max(axis=1)
country_min = non_zero_pivot_table.min(axis=1)

# Check if max equals min for countries with at least one non-NaN value
constant_index_countries = (country_max == country_min) & (country_max.notna())
print("\nPaíses con índice constante (sin considerar 0s como datos):")
if constant_index_countries.any():
    print(pivot_table_df[constant_index_countries].index.tolist())
else:
    print("Ningún país tiene un índice constante en los años registrados.")

# Pregunta e) ¿Qué países no tienen ningún dato (es decir, quedaron con todos los valores igual a 0)? ¿Podrías explicar por qué?
countries_all_zeros = pivot_table_df[(pivot_table_df == 0).all(axis=1)].index.tolist()
print("\nPaíses con todos los valores de 'indice' en 0 (sin datos registrados):")
print(countries_all_zeros)
print("Esto ocurre porque estos países no tenían registros en los archivos de libertad de prensa para los años cubiertos o sus 'indice' eran NaN y fueron reemplazados por 0 al crear la tabla dinámica. Es decir, no hay información para ellos en el dataset original o los valores eran nulos para todos los años.")
````
