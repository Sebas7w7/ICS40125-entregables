```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/ICS40125/main/docs/labs/data/netflix_titles.csv')
df.head()
```

## 📊 Parte 1: Exploración visual básica

1. **Distribución de tipos de contenido**

   * Realiza un gráfico de barras mostrando la cantidad de películas vs series (`type`).
   * Pregunta guía: ¿Cuál es el tipo de contenido predominante en Netflix? ¿Qué implicancias podría tener para los usuarios?

2. **Histograma de años de lanzamiento**

   * Muestra cómo se distribuyen los títulos según su `release_year`.
   * Pregunta guía: ¿Hay más contenido reciente o antiguo? ¿Notas algún patrón en décadas específicas?

3. **Proporción de clasificaciones por edad (`rating`)**

   * Haz un gráfico circular o de barras.
   * Pregunta guía: ¿Qué clasificación es más común? ¿Crees que Netflix está más enfocado en público adulto, familiar o infantil?



```python
# 1. Distribución de tipos de contenido
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Distribución de Tipos de Contenido en Netflix')
plt.xlabel('Tipo')
plt.ylabel('Cantidad de Títulos')
plt.show()

# Pregunta guía: ¿Cuál es el tipo de contenido predominante en Netflix? ¿Qué implicancias podría tener para los usuarios?
# Respuesta: Las películas son el tipo de contenido predominante. Esto podría implicar que Netflix atrae a usuarios que prefieren películas o que la plataforma invierte más en este tipo de contenido.

# 2. Histograma de años de lanzamiento
plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=30, kde=True, palette='plasma')
plt.title('Distribución de Títulos por Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Títulos')
plt.show()

# Pregunta guía: ¿Hay más contenido reciente o antiguo? ¿Notas algún patrón en décadas específicas?
# Respuesta: Hay una clara tendencia hacia contenido más reciente, con un pico significativo en los años 2010 y 2020, lo que sugiere una expansión reciente del catálogo.

# 3. Proporción de clasificaciones por edad (rating)
plt.figure(figsize=(10, 7))
df['rating'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
plt.title('Proporción de Clasificaciones por Edad en Netflix')
plt.ylabel('') # Eliminar la etiqueta 'type' del eje y
plt.show()

# Pregunta guía: ¿Qué clasificación es más común? ¿Crees que Netflix está más enfocado en público adulto, familiar o infantil?
# Respuesta: Las clasificaciones 'TV-MA' y 'TV-14' son las más comunes, lo que indica un enfoque significativo en el público adulto y adolescente. Las clasificaciones para niños ('TV-Y', 'TV-Y7', 'TV-G') y familiares ('PG', 'PG-13') también están presentes, pero en menor proporción, lo que sugiere una oferta diversa pero con un sesgo hacia contenido maduro.
```



## 🎨 Parte 2: Tendencias y evolución en el tiempo

4. **Número de títulos agregados por año**

   * Usa `date_added` (convertido a fecha) y grafica una serie temporal.
   * Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?

5. **Heatmap de lanzamientos por año y mes**

   * Construye un mapa de calor con `year_added` y `month_added`.
   * Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?

6. **Duración de películas por género**

   * Extrae la duración en minutos y crea un boxplot agrupado por género principal.
   * Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?


```python
# Preprocesamiento para Parte 2
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# 4. Número de títulos agregados por año
plt.figure(figsize=(12, 6))
df.groupby('year_added').size().plot(kind='line', marker='o')
plt.title('Número de Títulos Agregados a Netflix por Año')
plt.xlabel('Año')
plt.ylabel('Cantidad de Títulos Agregados')
plt.grid(True)
plt.show()

# Pregunta guía: ¿Cuándo creció más el catálogo de Netflix? ¿Qué relación podría tener con la expansión internacional de la plataforma?
# Respuesta: El catálogo de Netflix muestra un crecimiento exponencial a partir de 2014-2015, alcanzando su pico alrededor de 2019-2020. Esto coincide con el período de expansión global agresiva de la plataforma.

# 5. Heatmap de lanzamientos por año y mes
monthly_counts = df.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)
plt.figure(figsize=(15, 8))
sns.heatmap(monthly_counts, cmap='viridis', fmt='d', linewidths=.5, linecolor='lightgray')
plt.title('Heatmap de Títulos Agregados a Netflix por Año y Mes')
plt.xlabel('Mes')
plt.ylabel('Año')
plt.show()

# Pregunta guía: ¿Hay meses con más estrenos que otros? ¿Qué conclusiones puedes sacar sobre la estacionalidad de lanzamientos?
# Respuesta: El heatmap muestra que los últimos meses del año (octubre, noviembre, diciembre) y principios de año (enero) suelen tener un mayor número de adiciones de contenido, especialmente en los años de mayor crecimiento. Esto podría indicar una estrategia de lanzamiento de contenido para las temporadas de vacaciones o para el inicio de nuevos ciclos de consumo.

# 6. Duración de películas por género
movies_df = df[df['type'] == 'Movie'].copy()
# Extraer duración en minutos y manejar valores nulos/errores
movies_df['duration_minutes'] = movies_df['duration'].str.extract('(\d+)').astype(float)

# Simplificar los géneros tomando el primero si hay varios
movies_df['main_genre'] = movies_df['listed_in'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'Unknown')

# Filtrar géneros con pocas películas para una mejor visualización
genre_counts = movies_df['main_genre'].value_counts()
top_genres = genre_counts[genre_counts > 50].index # Seleccionar géneros con más de 50 películas
movies_df_filtered = movies_df[movies_df['main_genre'].isin(top_genres)]

plt.figure(figsize=(14, 8))
sns.boxplot(data=movies_df_filtered, x='main_genre', y='duration_minutes', palette='tab10')
plt.title('Distribución de Duración de Películas por Género Principal')
plt.xlabel('Género Principal')
plt.ylabel('Duración (minutos)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Qué géneros suelen durar más? ¿Qué podrías decir sobre la variabilidad dentro de un mismo género?
# Respuesta: Los géneros como 'Documentaries', 'Dramas' e 'Independent Films' suelen tener duraciones medianas más altas, mientras que géneros como 'Children & Family Movies' o 'Action & Adventure' pueden tener rangos más amplios. Se observa una considerable variabilidad en la duración dentro de casi todos los géneros, lo que indica que no hay una duración estricta para ningún tipo de película, aunque hay tendencias.
```



## 🔥 Parte 3: Comparaciones y relaciones

7. **Top 10 países con más producciones**

   * Usa un gráfico de barras para mostrar los países más frecuentes en la columna `country`.
   * Pregunta guía: ¿Qué países dominan el catálogo? ¿Sorprende la distribución o es esperada?

8. **Películas vs Series según género**

   * Realiza un gráfico de barras apiladas para mostrar la cantidad de títulos por género, separando `Movie` y `TV Show`.
   * Pregunta guía: ¿Existen géneros más asociados a películas o a series?

9. **Relación entre duración y año de lanzamiento**

   * Haz un gráfico de dispersión (scatterplot) con `release_year` en el eje X y `duration` (en minutos) en el eje Y.
   * Pregunta guía: ¿Han cambiado los estándares de duración de las películas con el tiempo? ¿Se ven outliers interesantes?



```python
# 7. Top 10 países con más producciones
# Limpiar y contar los países, ya que la columna 'country' puede contener múltiples países
# Para este análisis, consideraremos solo el primer país si hay una lista.
df['first_country'] = df['country'].apply(lambda x: str(x).split(',')[0].strip() if pd.notna(x) else 'Unknown')
top_countries = df['first_country'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='crest')
plt.title('Top 10 Países con Más Producciones en Netflix')
plt.xlabel('País')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Qué países dominan el catálogo? ¿Sorprende la distribución o es esperada?
# Respuesta: Estados Unidos domina abrumadoramente el catálogo, seguido por India, Reino Unido y Canadá. Esta distribución es en gran medida esperada, dado el tamaño del mercado de producción de contenido de EE. UU. e India, así como la expansión global de Netflix.

# 8. Películas vs Series según género
# Asegurarse de que 'listed_in' es string y manejar nulos
df_genres = df.copy()
df_genres['listed_in'] = df_genres['listed_in'].fillna('Unknown')
df_genres['main_genre'] = df_genres['listed_in'].apply(lambda x: x.split(',')[0].strip())

genre_type_counts = df_genres.groupby(['main_genre', 'type']).size().unstack(fill_value=0)
# Seleccionar los géneros principales para mejor visualización
top_genres_for_comparison = genre_type_counts.sum(axis=1).nlargest(15).index
genre_type_counts_filtered = genre_type_counts.loc[top_genres_for_comparison]

genre_type_counts_filtered.plot(kind='bar', stacked=True, figsize=(14, 7), cmap='coolwarm')
plt.title('Cantidad de Títulos por Género Principal (Películas vs. Series)')
plt.xlabel('Género Principal')
plt.ylabel('Cantidad de Títulos')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Contenido')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Existen géneros más asociados a películas o a series?
# Respuesta: 'Dramas', 'Comedies' y 'Action & Adventure' muestran un equilibrio entre películas y series o una ligera inclinación hacia películas. Sin embargo, géneros como 'International TV Shows', 'Crime TV Shows' y 'Kids'' TV' están fuertemente asociados con las series, mientras que 'Documentaries' y 'Children & Family Movies' están más asociados con películas.

# 9. Relación entre duración y año de lanzamiento
# Reutilizar movies_df del paso 6 para la duración en minutos
movie_duration_df = df[df['type'] == 'Movie'].copy()
movie_duration_df['duration_minutes'] = movie_duration_df['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(14, 7))
sns.scatterplot(data=movie_duration_df, x='release_year', y='duration_minutes', alpha=0.6, hue='rating', size='duration_minutes', sizes=(20, 400), palette='viridis')
plt.title('Relación entre Duración de Películas y Año de Lanzamiento')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Duración (minutos)')
plt.grid(True)
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Pregunta guía: ¿Han cambiado los estándares de duración de las películas con el tiempo? ¿Se ven outliers interesantes?
# Respuesta: No hay una tendencia clara y drástica en los estándares de duración a lo largo del tiempo; la mayoría de las películas se mantienen dentro de un rango de 70 a 120 minutos. Sin embargo, se pueden observar algunas películas de muy corta o muy larga duración (outliers) distribuidas a lo largo de los años, lo que sugiere que siempre ha habido espacio para la experimentación en la duración.
```
